# Setup
Imports and Japanese font support for Matplotlib.

```python
from pathlib import Path
from typing import List

import pandas as pd

COLUMN_MAP = {
    "商品名": "product",
    "説明": "description",
    "方向": "direction",
    "約定日": "trade_date",
    "対象期間(自)": "period_start",
    "対象期間(至)": "period_end",
    "売り入札量(MW)": "offer_mw",
    "約定価格(円/kWh)": "settlement_price_yen_per_kwh",
    "約定量(MW)": "settled_mw",
}

DIRECTION_MAP = {
    "東北 -> 北海道": "Tohoku -> Hokkaido",
    "東京 -> 中部": "Tokyo -> Chubu",
    "中部 -> 東京": "Chubu -> Tokyo",
    "四国 -> 関西": "Shikoku -> Kansai",
    "四国 -> 中国": "Shikoku -> Chugoku",
    "九州 -> 中国": "Kyushu -> Chugoku",
}

DATE_COLS = ["trade_date", "period_start", "period_end"]
NUM_COLS = ["offer_mw", "settlement_price_yen_per_kwh", "settled_mw"]


def _coerce_numeric(series: pd.Series) -> pd.Series:
    # Handle '-', empty, and strings; remove commas and spaces
    s = (
        series.astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
        .replace({"-": None, "": None})
    )
    return pd.to_numeric(s, errors="coerce")


def load_transmission_rights(raw_dir_path: str | Path) -> pd.DataFrame:
    raw_dir = Path(raw_dir_path)
    files: List[Path] = sorted(raw_dir.glob("transmission_rights_*.csv"))
    if not files:
        raise FileNotFoundError(f"No files matched in {raw_dir}")

    frames: List[pd.DataFrame] = []
    for f in files:
        try:
            df = pd.read_csv(f, encoding="cp932")
        except UnicodeDecodeError:
            df = pd.read_csv(f, encoding="shift_jis")
        df["source_file"] = f.name
        # Rename columns if present
        existing_map = {k: v for k, v in COLUMN_MAP.items() if k in df.columns}
        if existing_map:
            df = df.rename(columns=existing_map)
        frames.append(df)

    out = pd.concat(frames, ignore_index=True)

    # Parse dates
    for c in DATE_COLS:
        if c in out.columns:
            out[c] = pd.to_datetime(out[c], errors="coerce")

    # Parse numerics
    for c in NUM_COLS:
        if c in out.columns:
            out[c] = _coerce_numeric(out[c])

    # Add English direction translation
    if 'direction' in out.columns:
        out['direction_en'] = out['direction'].map(DIRECTION_MAP)

    return out


# Run loader
transmission_df = load_transmission_rights(
    "/Users/hankehly/Projects/adhoc-energy-analytics/data/raw"
)
print(
    f"Loaded {transmission_df.shape[0]:,} rows from {transmission_df['source_file'].nunique()} files"
)
print(transmission_df.dtypes)
transmission_df.head()
```

```python
# Check if periods have gaps or continuous coverage
print("Checking for gaps in period coverage...")
print("="*80)

# Find overall date range
min_date = transmission_df['period_start'].min()
max_date = transmission_df['period_end'].max()
print(f"Overall period range: {min_date.date()} to {max_date.date()}")

# Create a full date range for the dataset
full_date_range = pd.date_range(start=min_date, end=max_date, freq='D')
print(f"Total days in range: {len(full_date_range):,}")

# Create a set of all days covered by any period
covered_days = set()
for _, row in transmission_df.iterrows():
    period_dates = pd.date_range(start=row['period_start'], end=row['period_end'], freq='D')
    covered_days.update(period_dates)

covered_days_sorted = sorted(covered_days)
print(f"Days covered by at least one period: {len(covered_days_sorted):,}")

# Find gaps (uncovered days)
uncovered_days = [d for d in full_date_range if d not in covered_days]
print(f"Days NOT covered by any period: {len(uncovered_days)}")

if uncovered_days:
    print("\nGaps found:")
    # Group consecutive uncovered days
    gaps = []
    gap_start = uncovered_days[0]
    gap_end = uncovered_days[0]
    for day in uncovered_days[1:]:
        if (day - gap_end).days == 1:
            gap_end = day
        else:
            gaps.append((gap_start, gap_end))
            gap_start = day
            gap_end = day
    gaps.append((gap_start, gap_end))
    
    for gap_start, gap_end in gaps[:10]:  # Show first 10 gaps
        print(f"  {gap_start.date()} to {gap_end.date()} ({(gap_end - gap_start).days + 1} days)")
    if len(gaps) > 10:
        print(f"  ... and {len(gaps) - 10} more gaps")
else:
    print("\n✓ No gaps found! Every day is covered by at least one period.")
```

```python
# Analyze trade date frequency per direction
print("Trade date frequency by direction:")
print("="*80)

trade_dates_per_direction = transmission_df.groupby('direction')['trade_date'].agg([
    'nunique', 
    'min', 
    'max',
    lambda x: (x.max() - x.min()).days
]).rename(columns={'<lambda_0>': 'span_days'})

print(f"\nDirection-level trade date statistics:")
print(trade_dates_per_direction.sort_values('nunique', ascending=False))

# Calculate average days between consecutive trade dates per direction
print(f"\nAverage gap between consecutive trade dates by direction:")
print("-"*80)
for direction in sorted(transmission_df['direction'].unique()):
    dir_data = transmission_df[transmission_df['direction'] == direction].copy()
    unique_trade_dates = sorted(dir_data['trade_date'].unique())
    
    if len(unique_trade_dates) > 1:
        gaps = [(unique_trade_dates[i+1] - unique_trade_dates[i]).days for i in range(len(unique_trade_dates)-1)]
        avg_gap = sum(gaps) / len(gaps)
        min_gap = min(gaps)
        max_gap = max(gaps)
        print(f"  {direction}")
        print(f"    {len(unique_trade_dates)} unique trade dates over {trade_dates_per_direction.loc[direction, 'span_days']} days")
        print(f"    Gap between trades: min={min_gap}, max={max_gap}, avg={avg_gap:.1f} days")
    else:
        print(f"  {direction}: only 1 trade date")
```

```python
# Plot yearly offer_mw and settled_mw by direction for 2019-06 to 2024-06
import matplotlib.pyplot as plt
import numpy as np

# Ensure required columns exist
required_cols = {'direction', 'direction_en', 'period_start', 'offer_mw', 'settled_mw', 'source_file'}
missing = required_cols - set(transmission_df.columns)
if missing:
    raise ValueError(f"Missing columns for plotting: {missing}")

# Filter window
start_date = pd.Timestamp('2019-06-01')
end_date = pd.Timestamp('2024-06-30')
mask = (transmission_df['period_start'] >= start_date) & (transmission_df['period_start'] <= end_date)
plot_df = transmission_df.loc[mask].copy()

# Derive year from source file name (e.g., transmission_rights_2021.csv -> 2021)
plot_df['year'] = plot_df['source_file'].str.extract(r'(\d{4})', expand=False).astype(int)

# Aggregate per year and direction (English names) based on file year
agg = plot_df.groupby(['year', 'direction_en'], as_index=False)[['offer_mw', 'settled_mw']].sum()

# Overall aggregation per year
overall = agg.groupby('year', as_index=False)[['offer_mw', 'settled_mw']].sum()

# Plot: Overall
years = overall['year'].tolist()
x = np.arange(len(years))
width = 0.4

fig_overall, ax_overall = plt.subplots(figsize=(10, 4))
ax_overall.bar(x - width/2, overall['offer_mw'], width=width, color='steelblue', label='offer_mw')
ax_overall.bar(x + width/2, overall['settled_mw'], width=width, color='seagreen', label='settled_mw')
ax_overall.set_xticks(x)
ax_overall.set_xticklabels(years, rotation=0)
ax_overall.set_title('Yearly totals (2019-06 to 2024-06)')
ax_overall.set_ylabel('MW')
ax_overall.legend()
ax_overall.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# Subplots per direction (English names), in a grid (width=3)
directions = sorted(agg['direction_en'].unique())
cols = 3
rows = int(np.ceil(len(directions) / cols))
fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*3.2), sharey=True)
axes = np.array(axes).reshape(rows, cols)

for idx, direction in enumerate(directions):
    r = idx // cols
    c = idx % cols
    ax = axes[r, c]
    sub = agg[agg['direction_en'] == direction].copy()
    # Align years to full range for consistency
    sub = sub.set_index('year')
    full_years = years
    sub = sub.reindex(full_years).fillna(0)
    x = np.arange(len(full_years))
    ax.bar(x - width/2, sub['offer_mw'], width=width, color='steelblue')
    ax.bar(x + width/2, sub['settled_mw'], width=width, color='seagreen')
    ax.set_xticks(x)
    ax.set_xticklabels(full_years, rotation=0)
    ax.set_title(direction)
    ax.grid(axis='y', linestyle='--', alpha=0.5)

# Hide unused axes
for extra_idx in range(len(directions), rows*cols):
    r = extra_idx // cols
    c = extra_idx % cols
    axes[r, c].axis('off')

# Add shared legend
handles = [
    plt.Rectangle((0, 0), 1, 1, color='steelblue', label='offer_mw'),
    plt.Rectangle((0, 0), 1, 1, color='seagreen', label='settled_mw')
]
fig.legend(handles=handles, loc='upper center', bbox_to_anchor=(0.5, 1.04), ncol=2, frameon=False)

fig.suptitle('Yearly totals by direction (2019-06 to 2024-06)', y=0.99)
fig.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

```python
# Calculate annual average settlement prices
price_df = plot_df[plot_df['settled_mw'] > 0].copy()  # Only rows with actual settlements

# Simple average per year
simple_avg = price_df.groupby('year')['settlement_price_yen_per_kwh'].mean()

# Weighted average per year (weighted by settled_mw)
weighted_avg = price_df.groupby('year').apply(
    lambda x: (x['settlement_price_yen_per_kwh'] * x['settled_mw']).sum() / x['settled_mw'].sum()
)

print("Annual average settlement price (yen/kWh) - 2019-06 to 2024-06:")
print("="*80)
print("\nYear  | Simple Avg | Weighted Avg (by settled_mw)")
print("-"*55)
for year in sorted(simple_avg.index):
    print(f"{year}  | {simple_avg[year]:>10.2f} | {weighted_avg[year]:>10.2f}")
    
print("\n" + "="*80)
print(f"Overall simple average:   {simple_avg.mean():.2f} yen/kWh")
print(f"Overall weighted average: {weighted_avg.mean():.2f} yen/kWh")
```

```python
# Calculate average settlement prices by direction
price_by_dir = plot_df[plot_df['settled_mw'] > 0].copy()

# Simple average per direction
simple_avg_dir = price_by_dir.groupby('direction_en')['settlement_price_yen_per_kwh'].mean()

# Weighted average per direction (weighted by settled_mw)
weighted_avg_dir = price_by_dir.groupby('direction_en').apply(
    lambda x: (x['settlement_price_yen_per_kwh'] * x['settled_mw']).sum() / x['settled_mw'].sum(),
    include_groups=False
)

# Count of settlements per direction
settlement_counts = price_by_dir.groupby('direction_en').size()

print("Average settlement price by direction (yen/kWh) - 2019-06 to 2024-06:")
print("="*80)
print("\nDirection                | Simple Avg | Weighted Avg | # Settlements")
print("-"*75)
for direction in sorted(simple_avg_dir.index):
    print(f"{direction:24} | {simple_avg_dir[direction]:>10.2f} | {weighted_avg_dir[direction]:>12.2f} | {settlement_counts[direction]:>13,}")
    
print("\n" + "="*80)
print(f"Overall simple average:   {simple_avg_dir.mean():.2f} yen/kWh")
print(f"Overall weighted average: {weighted_avg_dir.mean():.2f} yen/kWh")
```

```python
# Calculate average settlement prices by direction and fiscal year
# Fiscal year = year from file name (already calculated in plot_df)

# Recompute plot_df if not available (aligns with earlier plotting filter)
if 'plot_df' not in globals():
    start_date = pd.Timestamp('2019-06-01')
    end_date = pd.Timestamp('2024-06-30')
    mask = (transmission_df['period_start'] >= start_date) & (transmission_df['period_start'] <= end_date)
    plot_df = transmission_df.loc[mask].copy()
    plot_df['year'] = plot_df['source_file'].str.extract(r'(\d{4})', expand=False).astype(int)

price_fy_df = plot_df[plot_df['settled_mw'] > 0].copy()

# Use the year column (from file name) as fiscal year
price_fy_df['fiscal_year'] = price_fy_df['year']

# Weighted average per direction and fiscal year
weighted_by_dir_fy = price_fy_df.groupby(['direction_en', 'fiscal_year']).apply(
    lambda x: (x['settlement_price_yen_per_kwh'] * x['settled_mw']).sum() / x['settled_mw'].sum(),
    include_groups=False
).reset_index(name='weighted_avg_price')

# Pivot to show years as columns
pivot_prices = weighted_by_dir_fy.pivot(index='direction_en', columns='fiscal_year', values='weighted_avg_price')

# Calculate overall weighted average per direction (across all years)
overall_by_direction = price_fy_df.groupby('direction_en').apply(
    lambda x: (x['settlement_price_yen_per_kwh'] * x['settled_mw']).sum() / x['settled_mw'].sum(),
    include_groups=False
)

print("Weighted average settlement price by direction and fiscal year (yen/kWh):")
print("="*110)
print("\nDirection                | FY2019 | FY2020 | FY2021 | FY2022 | FY2023 | FY2024 | Overall")
print("-"*110)
for direction in sorted(pivot_prices.index):
    row = f"{direction:24}"
    for year in range(2019, 2025):
        if year in pivot_prices.columns:
            val = pivot_prices.loc[direction, year]
            if pd.notna(val):
                row += f" | {val:6.2f}"
            else:
                row += " |    n/a"
        else:
            row += " |    n/a"
    overall_val = overall_by_direction[direction]
    row += f" | {overall_val:6.2f}"
    print(row)

print("\n" + "="*110)

# Calculate counts per direction and fiscal year
counts = price_fy_df.groupby(['direction_en', 'fiscal_year']).size().reset_index(name='count')
pivot_counts = counts.pivot(index='direction_en', columns='fiscal_year', values='count')

# Calculate overall counts per direction
overall_counts = price_fy_df.groupby('direction_en').size()

print("\nNumber of settlements by direction and fiscal year:")
print("="*110)
print("\nDirection                | FY2019 | FY2020 | FY2021 | FY2022 | FY2023 | FY2024 | Overall")
print("-"*110)
for direction in sorted(pivot_counts.index):
    row = f"{direction:24}"
    for year in range(2019, 2025):
        if year in pivot_counts.columns:
            val = pivot_counts.loc[direction, year]
            if pd.notna(val):
                row += f" | {val:6.0f}"
            else:
                row += " |    n/a"
        else:
            row += " |    n/a"
    overall_count = overall_counts[direction]
    row += f" | {overall_count:6.0f}"
    print(row)
```

```python
# Extract text elements from PDF page 12 to reconstruct the chart values
import pdfplumber
import re

pdf_path = "/Users/hankehly/Downloads/101_04_02.pdf"
page_index = 11

with pdfplumber.open(pdf_path) as pdf:
    page = pdf.pages[page_index]
    text = page.extract_text(x_tolerance=1, y_tolerance=1) or ""

print("Extracted text (page 12):")
print("=" * 80)
print(text[:4000])
print("=" * 80)

# Also show a quick sample of numeric tokens with approximate positions
words = page.extract_words(use_text_flow=True, keep_blank_chars=False)
num_re = re.compile(r"^\(?-?\d+(?:\.\d+)?\)?$")
nums = [w for w in words if num_re.match(w["text"])]
print(f"Found {len(nums)} numeric tokens.")
for w in nums[:60]:
    print(f"{w['text']:>8}  x0={w['x0']:.1f} top={w['top']:.1f}")
```

```python
# Recreate the page-12 chart (values transcribed from the rendered page image)
import numpy as np
import matplotlib.pyplot as plt

# The PDF chart shows two series per direction:
# - 平均発生値差 (Average spread) [blue]
# - 平均約定価格 (Average settlement price) [green]
# For FY, this notebook uses the same FY definition as the PDF page (2019年度..2023年度).

fys = [2019, 2020, 2021, 2022, 2023]

data = {
    "Tohoku → Hokkaido": {
        "avg_spread": [1.57, 0.42, -0.16, 0.05, 0.17],
        "avg_settle": [0.16, 0.29, 0.15, 0.01, 0.01],
        "ylim": (-0.5, 3.0),
    },
    "Tokyo → Chubu": {
        "avg_spread": [-1.85, -0.73, -0.38, -2.40, -1.05],
        "avg_settle": [0.01, 0.01, 0.01, 0.01, 0.01],
        "ylim": (-3.0, 0.5),
    },
    "Chubu → Tokyo": {
        "avg_spread": [1.85, 0.73, 0.38, 2.40, 1.05],
        "avg_settle": [1.18, 0.61, 0.01, 0.95, 0.28],
        "ylim": (-0.5, 3.0),
    },
    "Shikoku → Kansai": {
        "avg_spread": [0.03, 0.00, 0.04, 0.43, 0.24],
        "avg_settle": [0.01, 0.01, 0.01, 0.02, 0.04],
        "ylim": (-0.5, 3.0),
    },
    "Shikoku → Chugoku": {
        "avg_spread": [0.03, 0.01, 0.02, 0.10, 0.18],
        "avg_settle": [0.01, 0.01, 0.01, 0.01, 0.01],
        "ylim": (-0.5, 3.0),
    },
    "Kyushu → Chugoku": {
        "avg_spread": [0.35, 0.42, 2.95, 4.51, 0.53],
        "avg_settle": [0.06, 0.01, 1.26, 2.26, 0.33],
        "ylim": (-0.5, 5.0),
    },
}

# Layout matches the PDF (2 rows x 3 columns)
panels = [
    (0, 0, "Tohoku → Hokkaido"),
    (0, 1, "Tokyo → Chubu"),
    (0, 2, "Chubu → Tokyo"),
    (1, 0, "Shikoku → Kansai"),
    (1, 1, "Shikoku → Chugoku"),
    (1, 2, "Kyushu → Chugoku"),
]

fig, axes = plt.subplots(2, 3, figsize=(7.44, 5.6))
axes = np.array(axes).reshape(2, 3)

blue = "#3b6fb6"
green = "#3aa655"

for r, c, title in panels:
    ax = axes[r, c]
    sub = data[title]
    x = np.arange(len(fys))
    w = 0.32

    b = ax.bar(x - w / 2, sub["avg_spread"], width=w, color=blue, label="Average Spread")
    g = ax.bar(x + w / 2, sub["avg_settle"], width=w, color=green, label="Average Settlement Price")

    ax.set_title(title)
    # Only show y-axis label for left-most plots
    if c == 0:
        ax.set_ylabel("yen/kWh")
    else:
        ax.set_yticklabels([])
    ax.set_xticks(x)
    ax.set_xticklabels([f"FY{fy % 100}" for fy in fys])
    ax.set_ylim(*sub["ylim"])
    ax.grid(axis="y", color="#cccccc", linewidth=1, alpha=0.6)

    # Annotate values (negative in red parentheses like the PDF)
    for rect, val in zip(b, sub["avg_spread"]):
        if val < 0:
            txt = f"({abs(val):.2f})"
            color = "red"
            va = "top"
            y = rect.get_height() - 0.08
        else:
            txt = f"{val:.2f}"
            color = "black"
            va = "bottom"
            y = rect.get_height() + 0.06
        ax.text(rect.get_x() + rect.get_width() / 2, y, txt, ha="center", va=va, fontsize=9, color=color)

    for rect, val in zip(g, sub["avg_settle"]):
        txt = f"{val:.2f}"
        ax.text(rect.get_x() + rect.get_width() / 2, rect.get_height() + 0.06, txt, ha="center", va="bottom", fontsize=9, color="black")

# Shared legend moved away from the plots to avoid overlap
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", bbox_to_anchor=(0.5, 1.02), ncol=2, frameon=True)

# fig.suptitle("Average Spread and Average Settlement Price for Bilateral Transmission Rights (FY2019-FY2023)")
fig.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
